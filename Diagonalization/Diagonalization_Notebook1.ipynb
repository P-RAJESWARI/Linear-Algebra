T:R^2->R^2

R^2 : Two Dimensional Real line

Hence T represents the transformation from 2-D space to 2-D space

**Basis Vector:**
  With help of the basis vector,we can represent any other vector as linear combination of that basis vector in axis in which that vector acts as a basis

 ** Universal Basis:**
   With Universal basis,we can derive all the vector in any axis ie.
   We can represent all the vector as linear combination of the universal basis

   The Universal basis for 2-D space is

   e1=[[1],[0]]    e2=[[0],[1]]
   
   T=Dx

   x is the basis vector which is the basis in 2-D space

   T - the transformation of x with D is also a basis in 2-D space

   



```python
#import neccessary libraries to perform mathematical calculations on matrix,we import numpy and to observe the basis and its transformation,we import matplot
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Basis vector e1 and e2
#This is universal basis in 2-D space ,so we identify it as basis by just seeing
#Other vector can be identified as basis vector if those two vectors are linearly independent that is one vector cannot be expressed as linear combination of other
e1=np.array([[1],[0]])
e2=np.array([[0],[1]])
```

```python
def plotvector(t,e,lab1,lab2):
 # Create a figure and axis
 fig, ax = plt.subplots()

 # Plot the basis vectors
 ax.quiver(0, 0, t[0], t[1], angles='xy', scale_units='xy', scale=1, color='r', label=lab1)
 ax.quiver(0, 0, e[0], e[1], angles='xy', scale_units='xy', scale=1, color='b', label=lab2)

 # Set axis limits and labels
 ax.set_xlim([-1, 7])
 ax.set_ylim([-1, 7])
 ax.set_xlabel('X-axis')
 ax.set_ylabel('Y-axis')

 # Add a legend
 plt.grid()
 ax.legend()
```

```python
#We plot the basis vector ,they point 1 unit in x axis and 1 unit in y axis ,they exactly lie on the axis itself of positive x and y axis
plotvector(e1,e2,"e1","e2")
```

```python
def matrix_vector_multiply(A, v):
    # Get the number of rows and columns in matrix A
    arow = len(A)
    acol = len(A[0])
    # Get the length of vector v
    veclen = len(v)

    # Check if the number of columns in A is equal to the length of v for multiplication
    if acol != veclen:
        print("Multiplication cannot be done")
        return

    # Initialize the result vector with zeros
    result = [0 for _ in range(acol)]

    # Perform matrix-vector multiplication
    for i in range(arow):  # Loop over rows of A
        for j in range(acol):  # Loop over columns of A
            result[i] += A[i][j] * v[j]  # Accumulate the product of A[i][j] and v[j] in result[i]

    return np.array(result)
```

```python
#The diagonalization matrix which means we have non-zero values only in the diagonal position
D=np.array([[3,0],
            [0,6]])
```

```python
#Transforming the basis vector e1 by multiplying e1 with Diagonal matrix D ,T1=D*e1
T1=matrix_vector_multiply(D,e1)
print("Transformed e1 with diagonal matrix D:\n")
print(T1)
```

```python
#Transforming the basis vector e1 by multiplying e1 with Diagonal matrix D ,T2=D*e2
T2=matrix_vector_multiply(D,e2)
print("Transformed e2 with diagonal matrix D:\n")
print(T2)
```

```python
#Plotting the transformed basis vector
plotvector(T1,T2,"Transformed e1","Transformed e2")
```

The conclusion we may came about observing the transformed basis vector is that the vectors doesn't change its direction since we have only positive values in the diagonal position of Diagonal matrix and also the basis vectors are scaled by units that is the values in diagonal. The e1 is scaled by 6 units and e2 is scaled by 3 units

```python
#Let's have some vector x
x=np.array([[1],[1]])
```

```python
#We are transforming the vector by multiplying with the Diagonal matrix T=D*x
T=matrix_vector_multiply(D,x)
print(T)
```

```python
#Plotting the original vector x and the vector x after transformation to observe the transformation
plotvector(x,T,"x","Transformed x")
```

Here we can conclude ,that the vector x is also transformed by the same units ,that we have both values of vector is non-zero,the first value get scaled by 6 units and the second value gets scaled by 3 units

```python
#Squaring the diagonal matrix to observe the usage of Diagonal matrix
D_square=D*D
print(D_square)
```

While squaring diagonal matrix D,we can observe that the diagonal elements exactly squared,and it is also the diagonal matrix D.Hence Diagonal matrix can be used for scaling
