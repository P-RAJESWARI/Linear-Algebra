**Steps to solve this problem**

**Initialize Variables and Data:**
- Define the probability matrix `P` representing the transition probabilities between pages.
- Define the initial state vector `x` representing the starting page.

**Simulation for Starting at Page 1:**
- Initialize `x` with the probability of starting at Page 1.
- Perform matrix-vector multiplication with `P` for 100 iterations using a loop.
- Print the resulting composition of `x` after 100 iterations.

**Simulation for Starting at Page 3:**
- Initialize `x` with the probability of starting at Page 3.
- Perform matrix-vector multiplication with `P` for 100 iterations using a loop.
- Print the resulting composition of `x` after 100 iterations.

**Understanding the Result:**
- The resulting `x` vector represents the probabilities of being on each page after 100 transitions according to the transition matrix `P`.
- This `x` vector is commonly referred to as the steady-state or equilibrium distribution vector, describing the long-term behavior of the system.

**Function for multiplication of the matrix and the vector**

```python
def matrix_vector_multiply(A, v):
    # Get the number of rows and columns in matrix A
    arow = len(A)
    acol = len(A[0])
    # Get the length of vector v
    veclen = len(v)

    # Check if the number of columns in A is equal to the length of v for multiplication
    if acol != veclen:
        print("Multiplication cannot be done")
        return

    # Initialize the result vector with zeros
    result = [0 for _ in range(acol)]

    # Perform matrix-vector multiplication
    for i in range(arow):  # Loop over rows of A
        for j in range(acol):  # Loop over columns of A
            result[i] += A[i][j] * v[j]  # Accumulate the product of A[i][j] and v[j] in result[i]

    return result
```

The table represents the probability matrix (P) ie the columns represent the probability of landing on certain page(represented as rows) given that you are on certain page(column)

```python
#Defining transition matrix as column stochastic(sum of the values in each column is 1) matrix
transition_matrix=[[0,0,1,0.5],
                   [0.333333,0,0,0],
                   [0.333333,0.5,0,0.5],
                   [0.333333,0.5,0,0]]
```

Let's represent vector x as the probability of landing on a certain page
1. **Let's say you start on Page1,and then depending on the matrix ,you move to the next page.Repeat this process,say hundred times.How would the composition of the vector x look?**

```python
x=[1,0,0,0] #starting at the Page1,it's a deterministic event or sure event that the probability for sure event at Page1 is 1
# Assuming x is a vector representing the initial state
print("X initially when starting at page 1:", x)

# Perform matrix-vector multiplication with the transition matrix 100 times
for i in range(100):
    x = matrix_vector_multiply(transition_matrix, x)

print("Composition of vector X after multiplying with the transition matrix for 100 times:")
# Print the resulting vector representing the composition after 100 multiplications
for i, val in enumerate(x):
    print("Page {0} = {1}".format(i + 1, val))
```

**Takeway or Observation:** The composition of the vector X after multiplying with the transition matrix for 100 times ,we'll be able to figure out ,this is exactly the Eigen vector

 2.**Let's say you start at the page3,and then depending on the matrix ,you move to the next page.Repeat this process,say hundred times.How would the composition of the vector x look?**

```python
x=[0,0,1,0] #starting at the Page3,it's a deterministic event or sure event that the probability for sure event at Page3 is 1
# Assuming x is a vector representing the initial state
print("X initially when starting at page 1:", x)

# Perform matrix-vector multiplication with the transition matrix 100 times
for i in range(100):
    x = matrix_vector_multiply(transition_matrix, x)

print("Composition of vector X after multiplying with the transition matrix for 100 times:")
# Print the resulting vector representing the composition after 100 multiplications
for i, val in enumerate(x):
    print("Page {0} = {1}".format(i + 1, val))
```

3. **What is this x called?**

Irrespective of whether we start at the page1 or the page3 ,we get end up with the same vector,hence the vector X is known as steady state vector
