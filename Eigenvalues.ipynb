Problem Statement


Let A=[[1,6]
       [5,2]],

    u=[6,-5],
    v=[3,-2]
Are u and v eigen vectors of A? Prove that visually.

```python
#importing library numpy to do operation with matrix ie 2 dimensional numpy array
#importing library matplotlib.pyplot to work with different types of plots in python,here we use this to plot eigen vectors and matrix applied eigen vector to prove that the vector is eigen vector of the matrix as visually
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the matrix A as two dimensional numpy array
A = np.array([[1, 6],
              [5, 2]])
```

```python
# Define the eigenvectors as numpy array
u = np.array([6, -5])
v = np.array([3,-2])
```

```python
# Apply the matrix to the eigenvector
# This calculates the dot product of matrix A and the vector u resulting in new vector that matrix-vector product
Au = np.dot(A, u)
# This calculates the dot product of matrix A and the vector v resulting in new vector that matrix-vector product
Av = np.dot(A, v)
```

```python
# Define the starting point of the vectors (the origin)
origin = np.array([0, 0])

# Plot the first eigenvector 'u' in blue
plt.quiver(*origin, u[0], u[1], color='b', scale=5, label='Eigenvector 1')

# Plot the result of applying the matrix 'A' to 'u' (Au) in red
plt.quiver(*origin, Au[0], Au[1], color='r', scale=5, label='Matrix Applied Eigenvector 1')

# Plot the second eigenvector 'v' in green
plt.quiver(*origin, v[0], v[1], color='g', scale=5, label='Eigenvector 2')

# Plot the result of applying the matrix 'A' to 'v' (Av) in yellow
plt.quiver(*origin, Av[0], Av[1], color='y', scale=5, label='Matrix Applied Eigenvector 2')

# Set the limits for the x and y axes of the plot
plt.xlim(-5, 5)
plt.ylim(-5, 5)

# Label the x and y axes
plt.xlabel('x')
plt.ylabel('y')

# Add horizontal and vertical lines to represent the axes
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)

# Add a grid to the plot for better visualization
plt.grid(color='gray', linestyle='--', linewidth=0.5)

# Add a legend to explain the different vectors in the plot
plt.legend()

# Set the title of the plot
plt.title('Action of Matrix on Eigenvectors')

# Show the plot
plt.show()
```
